```python
def scroller(driver):
     """
    Scrolls a web page using a Selenium WebDriver.

    Args:
        driver: An instance of the Selenium WebDriver to interact with the browser.
    """
    scroll_pause_time = 1
    screen_height=driver.execute_script('return window.screen.height;')
    i=1
    while True:
        driver.execute_script('window.scrollTo(0,{screen_height}*{i});'.format(screen_height=screen_height,i=i))
        i+=1
        time.sleep(scroll_pause_time)
        scroll_height = driver.execute_script('return document.body.scrollHeight;')
        if (screen_height)*i>scroll_height*(3/4):
            break
```

```python
def concat_dataframe(dff1,repeated_df1):
    """
    Concatenates two DataFrames vertically and saves the resulting concatenated DataFrame to a CSV file.

    Args:
        dff1 (pandas.DataFrame): The first DataFrame to be concatenated.
        repeated_df1 (pandas.DataFrame): The second DataFrame to be concatenated, repeated multiple times.
    """
    dff1 = dff1.reset_index(drop = True)
    repeated_df1=repeated_df1.reset_index(drop = True)
    merged_df=pd.concat([repeated_df1,dff1],join='outer',axis=1)
    merged_df.to_csv('product_Camera.csv', index=False)
```
