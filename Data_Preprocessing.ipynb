# 4. Data Cleaning

## 4.1 Columns With Sum Of Null Values 

```python
df.isnull().sum()
```

## 4.2 Extracting Total Number Of Ratings Given Per Product Using extract

```python
df['Customer Reviews'].unique()
```

```python
df['Customer Reviews']=df['Customer Reviews'].str.extract('(\d+)\sratings')
df['Customer Reviews'].unique()
```

## 4.3 Extracting Data From review data Column Using extract

```python
df['review_date']=df['review_date'].str.extract(r"(\d+\s\w+\s\d{4})")
```

```python
df['review_date']
```

```python
re_mp=['Maximum resolution', 'Optical Sensor Resolution']
for mp in re_mp:
    df[mp] = df[mp].astype(str).str.extract(r'(\d+)')
```

## 4.4 Making Data Clear For Further Processing Using replace 

```python
rep=['current_price','actual_price']
for re in rep:
    df[re] = df[re].str.replace(',', '')
```

```python
df['manufacturer_rating'] = df['manufacturer_rating'].str.replace('%', '')
df['Digital zoom']=df['Digital zoom'].str.replace('x','')
df['Recording Capacity']=df['Recording Capacity'].str.replace('Minutes','')
```

## 4.5 Providing Meaningful Name To Columns

```python
df.rename(columns = {'manufacturer_rating':'manufacturer_rating in %', 'Digital zoom':'Digital zoom in x',
                     'Recording Capacity':'Recording Capacity in minutes'}, inplace = True)
```

```python
df
```

# 5. Data Reduction

```python
df.shape
```

## 5.1 Drop Columns Having Single Unique Value

```python
for col in df.columns:
    if df[col].nunique() == 1:
        df=df.drop(col, axis=1)
# df.nunique()
for col in df.columns:
    print(col,'-->',df[col].nunique())
```

## 5.2 Dropping Columns Which Are Not Required For Further Process

```python
# Blue color for Column name for highlighting
color_start = '\033[94m'
bold_start = '\033[1m'
bold_end = '\033[0m'
for i in df.columns:
    print(color_start+bold_start + i + bold_end,'-->',df[i].unique())
```

```python
# Create a list named dr for drop purpose, which contains column which have value count sum less than or equal to 1200 as it will have less impact on 
#  prediction as less data is there
dr=[]
for i in df.columns:
    val=df[i].value_counts().sum()
    if val <=1200:
        # print(i)
        dr.append(i)
print(dr)
```

```python
# Create a list named not_necess_drop_col for drop purpose as this are not important and make impact on prediction
not_necess_drop_col=['website','manufacturer','Manufacturer','Model','Model Year','Viewfinder type','ASIN','Best Sellers Rank','Date First Available','Item Dimensions LxWxH','Generic Name','Aspect Ratio','Minimum diaphragm opening','purchase_style','is_verified','Country of Origin']
not_necess_drop_col.extend(dr)   # extending list dr to list not_necess_drop_col
len(not_necess_drop_col)
```

```python
# Blue color for Column name for highlighting
color_start = '\033[94m'
bold_start = '\033[1m'
bold_end = '\033[0m'
for i in df.columns:
    print(color_start+bold_start + i + bold_end,'-->',df[i].unique())
```

```python
# Dropping the columns from dataframe
df=df.drop(not_necess_drop_col,axis=1)
df.shape
```
